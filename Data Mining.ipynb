#### 1. Handling Missing values

When one or more values are absent in a dataset, missing values occur. It is important to handle missing values for accurate analysis. Incorrect and incomplete data results in inaccurate and misleading analysis. 

To handle the missing values, we have some common techniques:

1. Drop rows with missing values:

This is the easiest approach, but it may result in a significant loss of data.

2. Fill Missing value with a constant:

This method replaced missing value with a constant value, such as mean, median or mode of the feature.

3. Impute missing values with machine learning technology:

This is an advanced techniques like k-nearesr neighbours (KNN) imputation, matrix factorisation, or even train in a mcahine learning model to predict the missing values (sidhikha, 2024).

#### 4. Data Discretisation
It is a data preprocessing step that is often used to transform a continuous-values attribute to a categorical attribute  (Koi.edu.au, 2024).


```python
# Assuming you have a DataFrame 'data_stroke' with columns: 'Age', 'BMI', 'Avg_glucose_level'

# Discretising 'age' into categories
Age_bins = [0, 20, 40, 60, 100]
Age_labels = ['0-20', '21-40', '41-60', '60+']
data_stroke['Age_group'] = pd.cut(data_stroke['Age'], bins=Age_bins, labels=Age_labels, right=False)

# Discretising 'bmi' into categories
BMI_bins = [0, 18.5, 24.9, 29.9, 40]
BMI_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
data_stroke['BMI_group'] = pd.cut(data_stroke['BMI'], bins=BMI_bins, labels=BMI_labels, right=False)

# Checking the updated DataFrame with discretized values
print(data_stroke[['Age', 'Age_group', 'BMI', 'BMI_group']].head())

#Visualisation

#Pie Chart for BMI Groups
BMI_counts = data_stroke['BMI_group'].value_counts()

plt.figure(figsize=(8, 8))
BMI_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Proportion of BMI Groups')
plt.ylabel('')  # Hides the y-axis label
plt.show()

#Box Plot for BMI Across Age Groups
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_group', y='BMI', data=data_stroke, palette='coolwarm')
plt.title('BMI Distribution Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('BMI')
plt.show()
```

#### Logistic Regression Model for Predicting Heart Disease with Preprocessing and Imputation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer

# Functioning to preprocess the datasets
def preprocess_data(data_stroke, data_heartdisease):
    # Converting 'Age' to float and standardize 'Gender' to ensure the datatypes of both datasets are same
    for data in [data_stroke, data_heartdisease]:
        data['Age'] = data['Age'].astype(float)
        data['Gender'] = data['Gender'].str.strip().str.lower()
    
    # Dropping rows with missing Age and Gender before merging
    data_stroke = data_stroke.dropna(subset=['Age', 'Gender'])
    data_heartdisease = data_heartdisease.dropna(subset=['Age', 'Gender'])

    # Merging the datasets on Age and Gender as Age and Gender are present in both datasets
    data_merged = pd.merge(data_stroke, data_heartdisease, on=['Age', 'Gender'], how='inner')
    
    if data_merged.shape[0] == 0:
        print("Warning: Merged dataset is empty. No matching rows found for Age and Gender.")
    
    return data_merged

# Preprocessing the datasets
data_merged = preprocess_data(data_stroke, data_heartdisease)

# Handling missing data by imputing numerical columns
imputer = SimpleImputer(strategy='mean')
data_merged[['Age', 'Avg_glucose_level', 'Hypertension']] = imputer.fit_transform(data_merged[['Age', 'Avg_glucose_level', 'Hypertension']])

# Encoding 'Gender' (if not already numeric)
encoder = LabelEncoder()
data_merged['Gender'] = encoder.fit_transform(data_merged['Gender'])

# Defining feature variables (X) and target variable (y)
X = data_merged[['Age', 'Gender', 'Hypertension', 'Avg_glucose_level']]
y = data_merged['HeartDisease']

# Logistic Regression modelling
model = LogisticRegression()
model.fit(X, y)

# Making predictions and evaluate the model
y_pred = model.predict(X)
print('Accuracy:', accuracy_score(y, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y, y_pred))
```

##### Key Observations:

Accuracy = 93.34%
Confusion Matrix:
    Tru Positives: 9603
    False Positives: 0
    False Negatives: 685
    True Negatives: 0

As the model is 93.34%, which suggests that the model is performing well. However, accuracy alone isn't enough to assess model performance when thee data is imbalanced.
